```python
def stanford_tokenize_and_map(text):
    text = text.lower()
    # Replace special Twitter elements with predefined tokens
    text = re.sub(r'@\w+', 'USER', text)  # Replace mentions with USER
    text = re.sub(r'#\w+', 'HASHTAG', text)  # Replace hashtags with HASHTAG
    text = re.sub(r'http\S+', 'URL', text)  # Replace URLs with URL
    tokens = word_tokenize(text)
    return ' '.join(tokens)
```

```python
undersampled_df['text'] = undersampled_df['text'].astype(str).apply(stanford_tokenize_and_map)
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

#Tokenization and Padding using Keras Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(undersampled_df['text'])
word_index = tokenizer.word_index
```

```python
import pickle

# Save the tokenizer to a pickle file
with open('tokenizer.pkl', 'wb') as f:
    pickle.dump(tokenizer, f)
```

```python
# Convert text to sequences
X = tokenizer.texts_to_sequences(undersampled_df['text'])
y = undersampled_df['label']
```

```python
# Pad sequences
max_sequence_length = 100  # Fixed length for input sequences
X_padded = pad_sequences(X, maxlen=max_sequence_length)
```

```python
def load_glove_embeddings(glove_file, embedding_dim):
    #Loads the GloVe embeddings from a file and returns a dictionary mapping words to vectors.
    embeddings_index = {}
    with open(glove_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    return embeddings_index
```

```python
# Load the GloVe embeddings (Twitter 100d)
glove_file = '/kaggle/input/cresci-2017-dataset/glove.6B.100d.txt' 
embedding_dim = 100
embeddings_index = load_glove_embeddings(glove_file, embedding_dim)
print(f"Loaded {len(embeddings_index)} word vectors from GloVe.")
```

```python
# Prepare Embedding Matrix
vocab_size = len(word_index) + 1  # Add 1 for padding index
embedding_matrix = np.zeros((vocab_size, embedding_dim))
```

```python
oov_token = np.random.normal(size=(embedding_dim,))  # Random vector for OOV words
```

```python
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        embedding_matrix[i] = embedding_vector  
    else:
        embedding_matrix[i] = oov_token  
```

# BI - DIRECTIONAL LSTM

```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.models import Sequential
from keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout, Input
from keras.optimizers import SGD
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, matthews_corrcoef
import re
import nltk
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
```

```python
# Step 6: Split Data into Training and Test Sets
X_train, X_test, y_train, y_test = train_test_split(X_padded, y, test_size=0.2, random_state=42)
```

```python
# Step 7: Build Three-Layer BiLSTM Model with GloVe Embeddings
model = Sequential()
```

```python
# Input layer specifying the shape of input data
model.add(Input(shape=(max_sequence_length,)))
```

```python
# Embedding layer (with pre-trained GloVe embeddings and trainable OOV token)
model.add(Embedding(input_dim=vocab_size, 
                    output_dim=embedding_dim, 
                    weights=[embedding_matrix], 
                    input_length=max_sequence_length, 
                    trainable=False))  # GloVe embeddings frozen
```

```python
# Three stacked BiLSTM layers with 200 units
model.add(Bidirectional(LSTM(200, return_sequences=True, input_shape=(max_sequence_length, embedding_dim))))  # First BiLSTM layer
model.add(Bidirectional(LSTM(200, return_sequences=True)))  # Second BiLSTM layer
model.add(Bidirectional(LSTM(200, return_sequences=False)))  # Third BiLSTM layer
```

```python
# Fully connected layer with dropout (Softmax will give posterior probabilities)
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
```

```python
# Output layer for binary classification (Softmax replaced with sigmoid for binary)
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the Model with SGD Optimizer
sgd = SGD(learning_rate=0.01, momentum=0.9)
model.compile(optimizer=sgd, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Model summary
model.summary()
```

```python
# Train the Model
history = model.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.2)
```

```python
# Train the Model
history = model.fit(X_train, y_train, epochs=5, batch_size=64, validation_split=0.2)
```

```python
# Evaluate the Model on Test Set
y_pred = model.predict(X_test).round().astype(int)
```

```python
# Calculate Evaluation Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
mcc = matthews_corrcoef(y_test, y_pred)
```

```python
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Matthews Correlation Coefficient (MCC): {mcc:.4f}")
```

```python
# Now you can use WordNetLemmatizer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import TweetTokenizer
import re
```

```python

# Function to preprocess tweets
def preprocess_tweets(tweets):
    lemmatizer = WordNetLemmatizer()
    tokenizer = TweetTokenizer()

 

    processed_tweets = []

    for tweet in tweets:

        # Step 1: Filter tweet markers (e.g., URLs, mentions)

        tweet = re.sub(r'http\S+|www\S+|https\S+', '', tweet, flags=re.MULTILINE)  # Remove URLs

        tweet = re.sub(r'@\w+', '', tweet)  # Remove mentions

        tweet = re.sub(r'#\w+', '', tweet)  # Remove hashtags

        # Step 2: Convert to lowercase

        tweet = tweet.lower()

        # Step 3: Remove duplicates (if any)

        tweet = ' '.join(sorted(set(tweet.split()), key=tweet.index))  # Remove duplicates while maintaining order

        # Step 4: Lemmatization

        words = tokenizer.tokenize(tweet)

        lemmatized_words = [lemmatizer.lemmatize(word) for word in words]

        processed_tweets.append(' '.join(lemmatized_words))

    return processed_tweets
```

```python
tweets = undersampled_df
```

```python
cleaned_tweets = preprocess_tweets(tweets)
```

```python
# Tokenization and Padding

tokenizer = Tokenizer()

tokenizer.fit_on_texts(cleaned_tweets)

sequences = tokenizer.texts_to_sequences(cleaned_tweets)
```

```python
# Padding sequences to ensure uniform input size

maxlen = max(len(seq) for seq in sequences)

padded_sequences = pad_sequences(sequences, maxlen=maxlen)
```

```python
# Assuming we have a mapping for language identification and embeddings (placeholders)

word_index = tokenizer.word_index
```

```python
# Define CNN-BiGRU Model

def create_model(embedding_matrix):
    input_layer = Input(shape=(max_sequence_length,))
    emb_layer = Embedding(input_dim=vocab_size, output_dim=embedding_dim, weights=[embedding_matrix],input_length=max_sequence_length, trainable=False)(input_layer)
    dropout1 = Dropout(0.5)(emb_layer)
    conv_layer = Conv1D(filters=128, kernel_size=5, activation='relu', kernel_regularizer='l2')(dropout1)
    batchnorm_layer = BatchNormalization()(conv_layer)
    act_layer = Activation('relu')(batchnorm_layer)
    bigru_layer = Bidirectional(GRU(64, return_sequences=True))(dropout1)  # BiGRU layer
    pooling_layer_conv = GlobalAveragePooling1D()(act_layer)
    pooling_layer_gru = GlobalAveragePooling1D()(bigru_layer)
    concatenate_layer = Concatenate()([pooling_layer_conv, pooling_layer_gru])
    dropout2 = Dropout(0.5)(concatenate_layer)
    output_layer = Dense(1, activation='sigmoid')(dropout2)  # Output layer
    model = Model(inputs=input_layer, outputs=output_layer)
    return model
```
